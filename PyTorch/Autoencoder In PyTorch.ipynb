```python
# Moving to GPU
device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
# If min and max is between 0 and 1, use `Sigmoid()`
transform = transforms.ToTensor()   #Transform the `Images` to the PyTorch Tensor

# If min and max is between 1 and 1, use `Tanh()`
# transform = transforms.Compose([
#     transforms.ToTensor(),
#     transforms.Normalize((0.5), (0.5))   # transforms.Normalize((First Dimention), (Second Dimention))
# ])

raw_mnist = datasets.MNIST(root='./', train=True, download=True, transform=transform)

data_loader = torch.utils.data.DataLoader(dataset=raw_mnist,
                                         batch_size=64,
                                         shuffle=True)

# epoch = 1 forward and backward pass of ALL training samples

# batch_size = number of training samples in one forward & backward pass

# number of iterations = number of passes, each pass using [batch_size] number of samples

# e.g. 100 samples, batch_size=20 --> 100/20 = 5 iterations for 1 epoch
```

```python
class Autoencoder_Linear(nn.Module):
    def __init__(self):
        # Size => N(batch _size), 784(28*28 is how many pixels in this `MNIST` datasets)
        super().__init__()
        # N, 784
        self.encoder = nn.Sequential(
            nn.Linear(28*28, 128),   # N, 784 -> N, 128   # (28*28 => Input Size, 128 => Reduced Size)
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 12),
            nn.ReLU(),
            nn.Linear(12, 3)   # N, 3
        )   
        
        # N,3 -> N, 784
        self.decoder = nn.Sequential(
            nn.Linear(3, 12),   # N, 784 -> N, 128   # (28*28 => Input Size, 128 => Reduced Size)
            nn.ReLU(),
            nn.Linear(12, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, 28*28),   # N, 3 -> N, 784
            nn.Sigmoid()   # Because we know the `images` have range between 0 to 1. So, `Sigmoid` activation function is suited
        ) 
    
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

# Note: [-1, 1] -> nn.Tanh()
# Note: [0, 1] -> nn.Sigmoid()
```

```python
model = Autoencoder_Linear().to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
```

```python
num_epochs = 10
outputs = []     # Create a list to store the outputs
for epoch in range(num_epochs):
    for (img, _) in data_loader:
        img = img.reshape(-1, 28*28).to(device)    # Because the images are in the shape (28, 28)
        recon = model(img)                         # we call the model to reconstruct the img
        loss = criterion(recon, img)               # Calculate the MSE
        
        optimizer.zero_grad()                      # Becareful to always zero_grad()
        loss.backward()
        optimizer.step()
    
    print(f'Epoch:{epoch+1}, Loss:{loss.item():.4f}')
    outputs.append((epoch, img, recon))
```

```python
for k in range(0, num_epochs, 4):
    plt.figure(figsize=(9, 2))
    plt.gray()
    imgs = outputs[k][1].detach().cpu().numpy()    # Convert to numpy array for ploting
    recon = outputs[k][2].detach().cpu().numpy()
    for i, item in enumerate(imgs):          # Plot the first nine images
        if i >= 9: break
        plt.subplot(2, 9, i+1)
        item = item.reshape(-1, 28, 28)      # Apply to reverse directions from (-1, 28*28) -> (-1, 28, 28)
        # item: 1, 28, 28
        plt.imshow(item[0])
        
    for i, item in enumerate(recon):
        if i >= 9: break
        plt.subplot(2, 9, 9+i+1)             # row_length + i + 1
        item = item.reshape(-1, 28, 28)
        # item: 1, 28, 28
        plt.imshow(item[0])
```

```python
class Autoencoder_CNN(nn.Module):
    def __init__(self):
        # Size => N(batch _size), 784(28*28 is how many pixels in this `MNIST` datasets)
        super().__init__()
        # N, 1, 28, 28 -> 
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 16, 3, stride=2, padding=1),   # (Input chanel, Output chanel, kernal size, stride, padding)-> (N, 16, 14, 14)
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),   # (N, 32, 7, 7)
            nn.ReLU(),
            nn.Conv2d(32, 64, 7)   # (N, 64, 1, 1)  # We increase the chanel but we reduces the image
        )   
        
        # (N, 64, 1, 1)
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 7),   #(N, 64, 1, 1) -> (N, 32, 7, 7)   
            nn.ReLU(),
            nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1),   #(N, 32, 7, 7) -> (N, 16, 14, 14)
            nn.ReLU(),
            nn.ConvTranspose2d(16, 1, 3, stride=2, padding=1, output_padding=1),    #(N, 16, 14, 14) -> (N, 1, 28, 28)
            nn.Sigmoid()   # Because we know the `images` have range between 0 to 1. So, `Sigmoid` activation function is suited
        ) 
    
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

# Note: [-1, 1] -> nn.Tanh()
# Note: [0, 1] -> nn.Sigmoid()
# nn.MaxPool2d -> nn.MaxUnpool2d
```

```python
model = Autoencoder_CNN().to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
```

```python
# Training Loops
num_epochs = 10
outputs = []     # Create a list to store the outputs
for epoch in range(num_epochs):
    for (img, _) in data_loader:
        recon = model(img.to(device))                        # we call the model to reconstruct the img
        loss = criterion(recon, img.to(device))              # Calculate the MSE
        
        optimizer.zero_grad()                      # Becareful to always zero_grad()
        loss.backward()
        optimizer.step()
    
    print(f'Epoch:{epoch+1}, Loss:{loss.item():.4f}')
    outputs.append((epoch, img, recon))
```

```python
for k in tqdm(range(0, num_epochs, 4)):
    plt.figure(figsize=(9, 2))
    plt.gray()
    imgs = outputs[k][1].detach().cpu().numpy()    # Convert to numpy array for ploting
    recon = outputs[k][2].detach().cpu().numpy()
    for i, item in enumerate(imgs):          # Plot the first nine images
        if i >= 9: break
        plt.subplot(2, 9, i+1)
        # item: 1, 28, 28
        plt.imshow(item[0])
        
    for i, item in enumerate(recon):
        if i >= 9: break
        plt.subplot(2, 9, 9+i+1)             # row_length + i + 1
        # item: 1, 28, 28
        plt.imshow(item[0])
```
