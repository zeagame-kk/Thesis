````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
from datetime import datetime
import pandas as pd
import plotly.express as px

from sklearn.ensemble import IsolationForest

mpl.rcParams['figure.figsize'] = (15, 8)
mpl.rcParams['axes.grid'] = False

import warnings
warnings.filterwarnings('ignore')
````

````python
df = pd.read_csv('datasets/nyc_taxi.csv')
df
````

````python
df.info()
````

````python
df['timestamp'] = pd.to_datetime(df['timestamp'], infer_datetime_format=True)
df.info()
````

````python
print(len(df))
df.head()
````

````python
df.tail()
````

````python
# Resample to 1H
df = df.set_index('timestamp').resample("H").mean().reset_index()
print(len(df))
df
````

````python
# Examine in one Month
fig, ax = plt.subplots(figsize=(15, 6))
df.plot(ax=ax, xlabel="timestamp", ylabel="value", title="NYC Taxi Demand");
````

````python
fig = px.line(df.reset_index(), x='timestamp', y='value', title='NYC Taxi Demand')

fig.update_xaxes(
    rangeslider_visible=True,
)
fig.show()
````

````python
df['hour']=df.timestamp.dt.hour
````

````python
df['weekday']=pd.Categorical(df.timestamp.dt.strftime('%A'),   # '%A' represents to Weekday as locale’s full name.
                             categories=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'],
                             ordered=True)
````

````python
df_final=df.join(df.groupby(['hour', 'weekday'])['value'].mean(), on=['hour','weekday'], rsuffix='_avg')
df_final
````

## Anomalous Point

```
NYC Marathon - 2014-11-02
Thanksgiving - 2014-11-27
Christmas - 2014-12-25
New Years - 2015-01-01
Snow Blizzard - 2015-01-26 and 2015-01-27
```

````python
# NYC Marathon
df_final.set_index('timestamp').loc['2014-11-02']
````

````python
# Snow Blizzard
df_final.set_index('timestamp').loc['2015-01-26':'2015-01-27']
````

````python
df_final['day'] = df.timestamp.dt.weekday
df_final
````

````python
data = df_final[['value', 'hour', 'day']]
data
````

# Train Model

````python
model =  IsolationForest(contamination=0.005, max_features=3, max_samples=0.8, n_estimators=200)
model.fit(data)
````

````python
df_final['outliers']=pd.Series(model.predict(data)).apply(lambda x: 'yes' if (x == -1) else 'no' )
````

````python
print("The anomaly events: ", len(df_final.query('outliers=="yes"')))
df_final.query('outliers=="yes"')
````

````python
fig = px.scatter(df_final, x='timestamp', y='value', color='outliers', hover_data=['weekday','hour','value_avg'], title='NYC Taxi Demand')

fig.update_xaxes(
    rangeslider_visible=True,
)
fig.show()
````

````python
score = model.decision_function(data)
score
````

````python
plt.hist(score, bins=50)
````

````python
df_final['scores'] = score
df_final
````

````python
print("The anomaly events: ", len(df_final.query('scores<-0.02')))
df_final.query('scores<-0.02')
````

````python
df_final['outliers_new'] = pd.Series(df_final['scores'] <= -0.02).apply(lambda x: 'yes' if (x == True) else 'no' )
df_final
````

````python
print("The anomaly events: ", len(df_final.query('outliers_new == "yes"')))
df_final.query('outliers_new == "yes"')
````

````python
fig = px.scatter(df_final, x='timestamp', y='value', color='outliers_new', hover_data=['weekday','hour','value_avg'], title='NYC Taxi Demand')

fig.update_xaxes(
    rangeslider_visible=True,
)
fig.show()
````
